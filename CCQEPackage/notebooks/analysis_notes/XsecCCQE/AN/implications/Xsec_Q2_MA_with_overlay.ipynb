```python
def find_x0_from_y0_in_y_array(y_poly , x_poly , y_0 , Epsilon=0.01):
    # return:
    # x_0 array of x's for which y(x)=y_0 within Epsilon
    # i_0 array of i's for which y(i)=y_0 within Epsilon
    x_0,i_0 = [],[]
    for i,(x,y) in enumerate(zip(x_poly,y_poly)):
        if np.abs(y-y_0)<Epsilon:
            x_0.append(x)
            i_0.append(i)  
    return np.array(x_0),np.array(i_0)
```

# cross-section as a function of $Q^2$

```python
debug=0
bins=linspace(0.,1.3,9)
vlabel=r'Q^{2}'; units='(GeV/c)^2'
wname = 'Pmu weight'
fig,ax = plt.subplots(figsize=(12,6))
mid = 0.5*(bins[1:]+bins[:-1]); bin_width=bins[1:] - bins[:-1]
var='reco_Q2_mcs'
Xsec_ctu_title = Xsec_ctu_titles[0]
Xsec_dict = get_Xsec_variable(debug=debug,
                              var=var,bins=bins,                              
                              wname=wname,                              
                              remove_ctu_bin=True,                              
                              selected_beam_on=selected_beam_on,selected_beam_off=selected_beam_off,                              
                              selected_overlay_concat=selected_overlay_concat,selected_CC1p=selected_CC1p)    
h = dict()
overlay_chi2 = dict()
h['Xsec'],h['Xsec err'] = Xsec_dict[var],Xsec_dict[var+' err']

ax.errorbar(x=mid,xerr=0.5*bin_width,y=h['Xsec'],yerr=h['Xsec err'],
            color=Colors['beam on'],fmt='o')        

h['mc Xsec'],h['mc Xsec err'] = Xsec_dict['mc '+var], Xsec_dict['mc '+var+' err']
ax.bar( x=mid , height=2*h['mc Xsec err'], bottom=h['mc Xsec']-h['mc Xsec err'],
       width=bin_width, color='blue')

genie_Xsec = dict()
genie_Xsec_err=dict()
chi2_data_arr = []
for mA,gname,glabel in zip(mA_array,genie_mA_list,genie_mA_labels):
    genie_mA = afro_genie_CC1p_dict[gname]
    # remove last cos(theta(mu)) bin
    genie_mA = genie_mA[genie_mA['truth_Pmu_cos_theta']<Bins['cos(theta(mu))'][-2]]
    h_genie,_ = np.histogram(genie_mA['truth_Q2'],bins=bins)
    h_genie_err = np.sqrt(h_genie)
    genie_Xsec[gname],genie_Xsec_err[gname] = np.zeros((len(bins)-1)), np.zeros((len(bins)-1))
    for i_bin in range(len(bins)-1):#{
        genie_Xsec[gname][i_bin],genie_Xsec_err[gname][i_bin] = compute_Xsec(Non=h_genie[i_bin]*4.908e19/4.9e20,  
                                                                   Non_err=h_genie_err[i_bin]*4.908e19/4.9e20,                                                                               
                                                                   bin_width=(bins[1]-bins[0]) )
    #}
    chi2_data,ndf = chi2_two_data_curves(h1=genie_Xsec[gname],h1err=genie_Xsec_err[gname],
                                        h2=h['Xsec'],h2err=h['Xsec err'],bins=bins,
                                        debug=0)
    overlay_chi2[mA],ndf = chi2_two_data_curves(h1=genie_Xsec[gname],h1err=genie_Xsec_err[gname],
                                        h2=h['mc Xsec'],h2err=h['mc Xsec err'],bins=bins,
                                        debug=0)
    chi2_data_arr.append(chi2_data)

    
    print "chi2[%.2f]=%.2f, chi2/ndf=%.2f"%(mA,chi2_data,chi2_data/ndf)
print 'done.'
```

```python
fig,ax = plt.subplots(figsize=(12,6))
x = mA_array
y = np.array(chi2_data_arr)/ndf
plt.plot(x,y,'o',color=Colors['beam on'])

par = np.polyfit(x,y,deg=4)
poly = np.poly1d(par)
x_fit = linspace(np.min(x),np.max(x),100)
plt.plot(x_fit,poly(x_fit),'--',color='royalblue')

set_axes(ax,r'$m_A$ [GeV/c$^2$]',r'$\chi^2/ndf$')
plt.tight_layout()
save_figure('/Users/erezcohen/Desktop/chi2_mA.pdf')
```

# What happens if we randomly sample events from the overlay instead of using the data?

```python
# first block: peal off CC1p from "overlay-concat"
selected_overlay_concat_no_CC1p = selected_overlay_concat[selected_overlay_concat.CC1p==False]
# print len(selected_overlay_concat_no_CC1p)
Ndata = len(selected_beam_on)
print 'simulating quasi-data for products of',Ndata,'data events'
```

```python
overlay_chi2_arr_dict = dict()
Nquasiexp = dict()

Nquasiexp['Ndata x 1'] = 20
Nquasiexp['Ndata x 4'] = 5

for Ndata_factor in [1,4]:
    key = 'Ndata x %d'%Ndata_factor
    print key,'\n----------'
    
    overlay_chi2_arr_dict[key] = dict()
    N_quasidata_CC1p = Ndata * Ndata_factor
    N_quasidata_nonCC1p = int((float(N_quasidata_CC1p)/len(selected_CC1p))*len(selected_overlay_concat_no_CC1p))    
    print N_quasidata_CC1p,'CC1p in quasidata, and',N_quasidata_nonCC1p,'non CC1p in quasidata'
    Normfact = float(len(selected_beam_on))/float(N_quasidata_CC1p)/Nevents['f(POT)']
    print 'Normfact:',Normfact

    
    for i_exp in range(Nquasiexp[key]):
        print 'quasi experiment',i_exp
        evnts_CC1p = np.sort(np.random.choice(range(len(selected_CC1p))
                                                     ,N_quasidata_CC1p
                                                     ,replace=False))
        quasidata_CC1p = selected_CC1p.iloc[evnts_CC1p]

        evnts_nonCC1p = np.sort(np.random.choice(range(len(selected_overlay_concat_no_CC1p))
                                                 ,N_quasidata_nonCC1p
                                                 ,replace=False))
        quasidata_nonCC1p = selected_overlay_concat_no_CC1p.iloc[evnts_nonCC1p]
        quasidata_concat = pd.concat([quasidata_nonCC1p,quasidata_CC1p])

        Xsec_dict = get_Xsec_variable(debug=debug,
                                      var=var,bins=bins,                              
                                      wname=wname,                              
                                      remove_ctu_bin=True,                              
                                      selected_beam_on=selected_beam_on,selected_beam_off=selected_beam_off,                              
                                      selected_overlay_concat=quasidata_concat,selected_CC1p=quasidata_CC1p)    
        h = dict()
        h['mc Xsec'],h['mc Xsec err'] = Xsec_dict['mc '+var], Xsec_dict['mc '+var+' err']
        # renormalize for the POT
        h['mc Xsec'] = h['mc Xsec']*Normfact
        h['mc Xsec err'] = h['mc Xsec err']*Normfact
        
        overlay_chi2_arr_dict[key][i_exp] = []
        for mA,gname in zip(mA_array,genie_mA_list):
            overlay_chi2,ndf = chi2_two_data_curves(h1=genie_Xsec[gname],
                                                              h1err=genie_Xsec_err[gname],
                                                              h2=h['mc Xsec'],
                                                              h2err=h['mc Xsec err'],bins=bins,
                                                              debug=0 if mA==1.1 else 0)
            overlay_chi2_arr_dict[key][i_exp].append(overlay_chi2)
    print 
print 'done.'    
```
