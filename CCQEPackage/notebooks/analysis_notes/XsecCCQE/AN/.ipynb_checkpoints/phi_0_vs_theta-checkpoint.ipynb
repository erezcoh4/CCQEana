# apply the efficiency weights

```python
def get_Xsecs(selected_beam_on,selected_beam_off,selected_overlay_concat,selected_CC1p,
              do_corr_phi_0=False, debug=0, do_proton=True, do_muon=True, do_P=True,
              do_cos_theta=True, do_phi=True, do_print_Xsec=True,
              remove_last_cos_theta_mu_bin=False):#{
    Xsec_dict = dict()
    for particle,do_particle in zip(['mu','p'],[do_muon,do_proton]):#{
        if do_particle==False: continue
        if debug: print 'getting cross-section with P'+particle+' weight'
        for i,(observable,ivar,do_var) in enumerate(zip(['P'+particle,'cos(theta('+particle+'))','phi('+particle+')']
                                                             ,[(1,4),(2,5),(3,6)]
                                                             ,[do_P,do_cos_theta,do_phi])):#{
            if do_var==False: continue
            var,vtrue,bins,mid,bin_width,vlabel,xlabel,units = get_labels(observable=observable)
            mul = 180./np.pi if 'phi' in observable else 1
            beam_on , beam_off , overlay , CC1p = selected_beam_on,selected_beam_off,selected_overlay_concat,selected_CC1p
            h = get_Xsec_1d(beam_on,beam_off,overlay,CC1p
                            ,var=var,bins=Bins[observable],bin_width=bin_width,wname='P'+particle+' weight',mul=mul
                            ,do_corr_phi_0=do_corr_phi_0)
            
            if debug>1:  pp.pprint(h)
            Xsec_dict[observable] = h['Xsec']
            Xsec_dict[observable+' err'] = h['Xsec err']
            Xsec_dict['mc '+observable] = h['mc Xsec']
            Xsec_dict['mc '+observable+' err'] = h['mc Xsec err']
        #}
    #}
    return Xsec_dict
```

```python
def draw_phi_Xsecs(Xsec_dict=None,filename='Xsec_Pmu_phi_Pp_phi_no_correction'
                   ,do_add_ellipses=False,do_save_figure=False,ylim_mu=None,ylim_p=None):
    fig=plt.figure(figsize=(20,8))
    for ip,(particle,iax,ylim) in enumerate(zip(['mu','p'],[(1,3),(2,4)],[ylim_mu,ylim_p])):
        print 'drawing cross-section with P'+particle+' weight'
        observable = 'phi('+particle+')'
        true = 'truth_P'+particle+'_phi'
        var,vtrue,bins,mid,bin_width,vlabel,xlabel,units = get_labels(observable=observable)
        mul = 180./np.pi
        h = dict()
        h['Xsec'],h['Xsec err'] = Xsec_dict[observable],Xsec_dict[observable+' err']
        h['mc Xsec'],h['mc Xsec err'] = Xsec_dict['mc '+observable], Xsec_dict['mc '+observable+' err']

        ax = fig.add_subplot(3,2,iax)
        plt.errorbar(x=mid,xerr=0.5*bin_width,y=h['Xsec'],yerr=h['Xsec err'],color=Colors['beam on'],fmt='o',label='data')
        ax.bar( x=mid , height=2*h['mc Xsec err'], bottom=h['mc Xsec']-h['mc Xsec err'], width=bin_width, color=Colors['CC1p'],label='overlay')
        if do_add_ellipses:
            if particle=='mu':
                e0 = patches.Ellipse((0, 0.012), 35, 0.02, color='red', linewidth=3, fill=False, zorder=2)
                ax.add_patch(e0)
            if particle=='p':
                e0 = patches.Ellipse((0, 0.011), 35, 0.013, color='red', linewidth=3, fill=False, zorder=2)
                ax.add_patch(e0)
                e_p_pi = patches.Ellipse((mid[0], 0.011), 30, 0.012, color='red', linewidth=3, fill=False, zorder=2)
                e_m_pi = patches.Ellipse((mid[-1], 0.011), 30, 0.012, color='red', linewidth=3, fill=False, zorder=2)
                ax.add_patch(e_p_pi)
                ax.add_patch(e_m_pi)
        set_axes(ax,'',y_label=get_Xsec_label(vlabel=vlabel,units=units)
                         ,do_add_grid=True,remove_ticks_x=True,do_add_legend=True if ip==1 else 0                                  
                         ,ylim=(0,1.1*np.max(ax.get_ylim())) if ylim is None else ylim
                )    
        # residuals plot
        den, den_err = h['Xsec']-h['mc Xsec'],np.sqrt(np.square(h['Xsec err'])+np.square(h['mc Xsec err']))
        num, num_err = h['Xsec'],h['mc Xsec err']
        ratio = den/num
        ax = fig.add_subplot(3,2,5+ip)
        ratio_err = ratio*np.sqrt(np.square(den_err/den) + np.square(num_err/num))
        
        plt.errorbar(x=mid,xerr=0.5*bin_width,y=ratio,yerr=ratio_err,fmt='o',markersize=0,color='black')
        chi2,ndf = chi2_two_data_curves(h1=h['Xsec'],h1err=h['Xsec err'],h2=h['mc Xsec'],h2err=h['mc Xsec err'],bins=bins,debug=0)
        plt.text(mid[0]-30,0.94*np.min(ax.get_ylim()) + 0.06*np.max(ax.get_ylim()), r'$\chi^2/ndf=%.2f/%d$'%(chi2,ndf), fontsize=20)
        set_axes(ax,xlabel,y_label=r'(data-MC)/data',do_add_grid=True)
        plt.plot([np.min(bins),np.max(bins)],[0,0],'--',color='royalblue')            
        
    plt.tight_layout(h_pad=0.0)   
    if do_save_figure:
        save_figure(figures_path+filename+'.pdf')
```

```python
Xsec_dict_all = get_Xsecs( selected_beam_on,selected_beam_off,selected_overlay_concat,selected_CC1p, do_corr_phi_0=False )
Xsec_dict_ctu_small = get_Xsecs( beam_on_ctu_small,beam_off_ctu_small,overlay_concat_ctu_small,CC1p_ctu_small, do_corr_phi_0=False)
Xsec_dict_ctu_large = get_Xsecs( beam_on_ctu_large,beam_off_ctu_large,overlay_concat_ctu_large,CC1p_ctu_large, do_corr_phi_0=False )
```

# look at the ratio of data to overlay in 2D ($\theta_\mu$ vs. $\phi_\mu$ and $\phi_p$ and $\theta_p$ vs. $\phi_p$)

```python
def get_Xsec_2D_dict(selected_beam_on,selected_beam_off,selected_overlay_concat,selected_CC1p,
                     bins_costheta_mu=None,bins_costheta_p=None,bins_phi=None):#{
    Xsec_dict = dict()
    for particle in ['mu','p']:#{
        print 'getting cross-section with P'+particle+' weight'
            
        var_x,vtrue_x,bins_x,mid_x,bin_width_x,vlabel_x,xlabel_x,units_x = get_labels(observable='cos(theta('+particle+'))')
        var_y,vtrue_y,bins_y,mid_y,bin_width_y,vlabel_y,xlabel_y,units_y = get_labels(observable='phi('+particle+')')        
        if particle=='mu':
            bins_ct = bins_costheta_mu if bins_costheta_mu is not None else bins_x 
        else:
            bins_ct = bins_costheta_p if bins_costheta_p is not None else bins_x 
            
        if bins_phi is None:  bins_ph = bins_y
        else: bins_ph = bins_phi
            
        mul_x = 1
        mul_y = 180./np.pi
        beam_on , beam_off , overlay , CC1p = selected_beam_on,selected_beam_off,selected_overlay_concat,selected_CC1p
        h = get_Xsec_2d(beam_on,beam_off,overlay,CC1p                        
                        ,var_x=var_x,bins_x=bins_ct,bin_width_x=(bins_ct[1:] - bins_ct[:-1])
                        ,var_y=var_y,bins_y=bins_ph,bin_width_y=(bins_ph[1:] - bins_ph[:-1])
                        ,wname='P'+particle+' weight'
                        ,mul_x=mul_x,mul_y=mul_y)            
        #}
        Xsec_dict['Xsec ' + particle] = h['Xsec']
        Xsec_dict['Xsec data/mc ' + particle] = h['Xsec data/mc']
    #}
    return Xsec_dict
```

```python
# ----------------------------------------------------------
# Jan-16, 2019
def get_Xsec_2d(beam_on=None,beam_off=None,overlay=None,CC1p=None,evtwgt_name='',
                var_x='reco_Pmu_cos_theta',var_y='reco_Pmu_phi',
                bins_x=Bins['cos(theta(mu))'],bin_width_x=None,
                bins_y=Bins['phi(mu)'],bin_width_y=None,
                wname='Pmu weight',
                mul_x=1,mul_y=1,
                debug=0):#{
    h=dict()
    for sam,slabel in zip([beam_on,beam_off,overlay,CC1p] ,['beam on','beam off','overlay','CC1p']):#{
        h[slabel],h[slabel+' err'] = np.zeros((len(bins_x)-1,len(bins_y)-1)),np.zeros((len(bins_x)-1,len(bins_y)-1))
        for i_x in range(len(bins_x)-1):#{
            sam_in_bin_x = sam[(bins_x[i_x] <= mul_x*sam[var_x])
                                 & (mul_x*sam[var_x] < bins_x[i_x+1])]
            if debug: print len(sam_in_bin_x),'events in ',bins_x[i_x],'<cos(theta)<',bins_x[i_x+1]
            
            for i_y in range(len(bins_y)-1):#{
                
                sam_in_bin = sam_in_bin_x[(bins_y[i_y] <= mul_y*sam_in_bin_x[var_y])
                                         & (mul_y*sam_in_bin_x[var_y] < bins_y[i_y+1])]
                
                if debug: print '\t',len(sam_in_bin),'events in ',bins_y[i_y],'<phi<',bins_y[i_y+1]
                
                weights_in_bin = sam_in_bin[wname].fillna(0)
                h[slabel][i_x][i_y] = np.sum( weights_in_bin )
                h[slabel+' err'][i_x][i_y] = np.sqrt(np.sum(np.square(sam_in_bin[wname+' err'].fillna(0))
                                                            +np.square(sam_in_bin[wname].fillna(0))))
            #}
        #}
    #}
    
    h['B'] = (h['overlay'] - h['CC1p'])*Nevents['f(POT)']
    h['B err'] = np.sqrt(np.square(h['overlay err']) + np.square(h['CC1p err']))*Nevents['f(POT)']
    
    h['N(on)'] = h['beam on']
    h['N(on) err'] = h['beam on err']
    
    h['N(off)'] = h['beam off']*OffBeam_scaling
    h['N(off) err'] = h['beam off err']*OffBeam_scaling
    
    h['N(on)-N(off)-B'] = h['N(on)'] - h['N(off)'] - h['B']
    h['N(on)-N(off)-B err'] = np.sqrt(np.square(h['N(on) err']) + np.square(h['N(off) err']) + np.square(h['B err']))
    
    
    h['Xsec'] = np.zeros((len(bins_x)-1,len(bins_y)-1)); 
    h['Xsec err'] = np.zeros((len(bins_x)-1,len(bins_y)-1));
    h['mc Xsec'] = np.zeros((len(bins_x)-1,len(bins_y)-1))
    h['mc Xsec err'] = np.zeros((len(bins_x)-1,len(bins_y)-1))
    h['Xsec data/mc'] = np.zeros((len(bins_x)-1,len(bins_y)-1))

    for i_x in range(len(bins_x)-1):#{
        for i_y in range(len(bins_y)-1):#{
            h['Xsec'][i_x][i_y] = h['N(on)-N(off)-B'][i_x][i_y]/(bin_width_x[i_x] * bin_width_y[i_y])
            h['Xsec err'][i_x][i_y] = h['N(on)-N(off)-B err'][i_x][i_y]/(bin_width_x[i_x] * bin_width_y[i_y])

            CC1p_in_bin = CC1p[(bins_x[i_x] <= mul_x*CC1p[var_x])
                                 & (mul_x*CC1p[var_x]<bins_x[i_x+1])
                                 & (bins_y[i_y] <= mul_y*CC1p[var_y])
                                 & (mul_y*CC1p[var_y]<bins_y[i_y+1])]
            
            h['mc Xsec'][i_x][i_y] = np.sum(CC1p_in_bin[wname])*Nevents['f(POT)']/(bin_width_x[i_x] * bin_width_y[i_y])
            h['mc Xsec err'][i_x][i_y] = h['CC1p err'][i_x][i_y]*Nevents['f(POT)']/(bin_width_x[i_x] * bin_width_y[i_y])
            
            h['Xsec data/mc'][i_x][i_y] = h['Xsec'][i_x][i_y] / h['mc Xsec'][i_x][i_y]
        #}
    #}
    
    return h
#}
# ----------------------------------------------------------

```

```python
bins_phi=linspace(-180,180,6)
bins_costheta_mu = linspace(-0.65,0.95,6)
bins_costheta_p = linspace(0.15,1,5)
Xsec_dict_2D = get_Xsec_2D_dict( selected_beam_on,selected_beam_off,selected_overlay_concat,selected_CC1p,
                                bins_costheta_mu=bins_costheta_mu,bins_costheta_p=bins_costheta_p,
                                bins_phi=bins_phi)
```
