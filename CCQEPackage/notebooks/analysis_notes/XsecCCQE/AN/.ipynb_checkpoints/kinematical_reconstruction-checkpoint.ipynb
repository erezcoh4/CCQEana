### Kinematical distributions of the selected events

```python
fig=plt.figure(figsize=(24,8))
for iax,(observable,truth_var
         ,mul,bins) in enumerate(zip(['Pp','cos(theta(p))','phi(p)'],
                                     ['truth_Pp','truth_Pp_cos_theta','truth_Pp_phi'],                                                    
                                     [1.,1.,180./np.pi],
                                     [linspace(0.25,1.15,50),linspace(0.2,1,50),linspace(-180,180,50)]                                     
                                    )):
    reco_var,vtrue,bins,mid,bin_width,vlabel,xlabel,units = get_labels(observable=observable)
    ax=fig.add_subplot(1,3,iax+1)
    plot_2d_hist(mul*selected_CC1p[truth_var],mul*selected_CC1p[reco_var]
                 ,xlabel='gen. ' + xlabel
                 ,ylabel='rec. ' + xlabel
                 ,bins=bins
                 ,ax=ax);
    ax.plot([np.min(bins),np.max(bins)],[np.min(bins),np.max(bins)],'--'
            ,color='white',alpha=1,linewidth=3)
plt.tight_layout()
save_figure(figures_path+'p_reconstruction.pdf')
```

```python
fig=plt.figure(figsize=(24,8))
for iax,(observable,truth_var
         ,mul,bins) in enumerate(zip(['Pmu','cos(theta(mu))','phi(mu)'],
                                     ['truth_Pmu','truth_Pmu_cos_theta','truth_Pmu_phi'],                                                    
                                     [1.,1.,180./np.pi],
                                     [linspace(0.1,1.2,50),linspace(-0.2,1,50),linspace(-180,180,50)]                                     
                                    )):
    reco_var,vtrue,bins,mid,bin_width,vlabel,xlabel,units = get_labels(observable=observable)

    ax=fig.add_subplot(1,3,iax+1)
    plot_2d_hist(mul*selected_CC1p[truth_var],mul*selected_CC1p[reco_var]
                 ,xlabel='gen. ' + xlabel
                 ,ylabel='rec. ' + xlabel
                 ,bins=bins
                 ,ax=ax);
    ax.plot([np.min(bins),np.max(bins)],[np.min(bins),np.max(bins)],'--'
            ,color='white',alpha=1,linewidth=3)
plt.tight_layout()
save_figure(figures_path+'mu_reconstruction.pdf')
```

# neutrino direction (Feb-14, 2019)

```python
reco_Pmu_x,reco_Pmu_y,reco_Pmu_z = selected_CC1p['reco_Pmu_mcs_x'],selected_CC1p['reco_Pmu_mcs_y'],selected_CC1p['reco_Pmu_mcs_z']
reco_Pp_x,reco_Pp_y,reco_Pp_z = selected_CC1p['reco_Pp_x'],selected_CC1p['reco_Pp_y'],selected_CC1p['reco_Pp_z']
reco_Pnu_x,reco_Pnu_y,reco_Pnu_z = reco_Pp_x + reco_Pmu_x, reco_Pp_y + reco_Pmu_y, reco_Pp_z + reco_Pmu_z
reco_Pnu = selected_CC1p['reco_Ev_mcs']
reco_Pnu_theta = np.arccos( reco_Pnu_z / reco_Pnu )

truth_Pmu_x,truth_Pmu_y,truth_Pmu_z = selected_CC1p['truth_Pmu_x'],selected_CC1p['truth_Pmu_y'],selected_CC1p['truth_Pmu_z']
truth_Pp_x,truth_Pp_y,truth_Pp_z = selected_CC1p['truth_Pp_x'],selected_CC1p['truth_Pp_y'],selected_CC1p['truth_Pp_z']
truth_Pnu_x,truth_Pnu_y,truth_Pnu_z = truth_Pp_x + truth_Pmu_x, truth_Pp_y + truth_Pmu_y, truth_Pp_z + truth_Pmu_z
truth_Pnu = selected_CC1p['truth_Ev']
truth_Pnu_theta = np.arccos( truth_Pnu_z / truth_Pnu )
```

```python
xlabel = r'$\theta_\nu$ [deg.]'
bins = linspace(0,70,25)
fig=plt.figure(figsize=(20,8))
ax=fig.add_subplot(1,2,1)
plot_2d_hist(180./np.pi*truth_Pnu_theta,180./np.pi*reco_Pnu_theta
                 ,xlabel='gen. ' + xlabel
                 ,ylabel='rec. ' + xlabel
                 ,bins=bins
                 ,ax=ax);
ax.plot([np.min(bins),np.max(bins)],[np.min(bins),np.max(bins)],'--'
            ,color='white',alpha=1,linewidth=3)
ax.set_title('2D correlation distribution',fontsize=25)

ax=fig.add_subplot(1,2,2)
x = 180./np.pi*(truth_Pnu_theta - reco_Pnu_theta)
x = x[x<180]
plt.hist(x,bins=linspace(-60,60,30));
set_axes(ax,x_label='gen. - rec. ' + xlabel,y_label='counts',title=r'1D difference distribution, $\sigma = %.1f^{\circ}$'%np.std(x))

plt.tight_layout()
save_figure(figures_path+'theta_v_reconstruction.pdf')
```

```python
overlay_genie = pd.read_csv('/Users/erezcohen/Desktop/uBoone/CCQEanalysis/csvFiles/genie/'
                                    +versions['overlay date']+'/'
                                    +versions['Overlay']+'_'+versions['overlay date']+'_genie.csv')
```

```python
print len(overlay_genie),'events in genie from overlay'
overlay_genie_CC1p = overlay_genie[(overlay_genie.IsCC_1p_200MeVc==True)
                                           & ((overlay_genie.truth_x>3) & (overlay_genie.truth_x<256))
                                           & ((overlay_genie.truth_y>-115) & (overlay_genie.truth_y<115))
                                           & ((overlay_genie.truth_z>5) & (overlay_genie.truth_z<1037))
                                           ]
```

```python
truth_Pmu_x,truth_Pmu_y,truth_Pmu_z = overlay_genie_CC1p['truth_Pmu_x'],overlay_genie_CC1p['truth_Pmu_y'],overlay_genie_CC1p['truth_Pmu_z']
truth_Pp_x,truth_Pp_y,truth_Pp_z = overlay_genie_CC1p['truth_Pp_x'],overlay_genie_CC1p['truth_Pp_y'],overlay_genie_CC1p['truth_Pp_z']
truth_Pv = overlay_genie_CC1p['truth_Ev']
truth_Pv_x_n_at_rest,truth_Pv_y_n_at_rest,truth_Pv_z_n_at_rest = truth_Pp_x + truth_Pmu_x, truth_Pp_y + truth_Pmu_y, truth_Pp_z + truth_Pmu_z

truth_Pv_n_at_rest = np.sqrt(np.square(truth_Pv_x_n_at_rest) + np.square(truth_Pv_y_n_at_rest) + np.square(truth_Pv_z_n_at_rest))
truth_Pv_theta_n_at_rest = np.arccos( truth_Pv_z_n_at_rest / truth_Pv_n_at_rest )
```

```python
fig=plt.figure(figsize=(24,8))
ax = fig.add_subplot(1,3,1)
plot_2d_hist(overlay_genie_CC1p['truth_Ev'],truth_Pv_n_at_rest,bins=linspace(0,2,100),ax=ax);
set_axes(ax,r'true $p^\nu$ [GeV/c]',r' $p^\nu$ for assuming n at rest [GeV/c]',xticks=(0,0.5,1,1.5),yticks=(0,0.5,1,1.5),do_add_grid=True)

ax = fig.add_subplot(1,3,2)
plot_2d_hist(overlay_genie_CC1p['truth_Pv_z'],truth_Pv_z_n_at_rest,bins=linspace(0,2,100),ax=ax);
set_axes(ax,r'true $p_z^\nu$ [GeV/c]',r' $p_z^\nu$ for assuming n at rest [GeV/c]',xticks=(0,0.5,1,1.5),yticks=(0,0.5,1,1.5),do_add_grid=True)



ax=fig.add_subplot(1,3,3)
plot_2d_hist( 180./np.pi*overlay_genie_CC1p['truth_Pv_theta'], 180./np.pi*np.arccos( truth_Pv_z_n_at_rest / truth_Pv_n_at_rest) 
             ,bins=(linspace(0.,0.3,100),linspace(0,60,100))
             ,ax=ax);
set_axes(ax,r'true $\theta^\nu$ [deg.]',r' $\theta^\nu$ for assuming n at rest [deg.]')

plt.tight_layout()
save_figure(figures_path+'theta_v_smearing_from_moving_neutron.pdf')
```
